
22\. Groundbreaking CNN Architectures
=====================================

### 

### Summary: Evolution of CNN Architectures for ImageNet Competition

#### 1. ImageNet and the Birth of Image Recognition Competitions:
- **ImageNet:** Database of 10 million labeled images across 1000 categories.
- **Competition:** Annual ImageNet Large Scale Visual Recognition Competition since 2010.



#### 2. Breakthroughs in CNN Architectures:
- **2012:** **AlexNet**
  - Developed by a team at the University of Toronto.
  - Pioneered ReLU activation function and dropout to prevent overfitting.
  
  
  
- **2014:** **VGGNet (VGG 16 & VGG 19)**
  - Developed by Oxford University's Visual Geometry Group.
  - Simple architecture with sequences of 3x3 convolutions, 2x2 pooling, and fully connected layers.
  
  
  
  
  
- **2015:** **ResNet**
  - Developed by Microsoft Research.
  - Addressed vanishing gradient problem by adding skip connections, enabling deep networks (up to 152 layers).
  
  
  


#### 3. Advancements in Deep CNNs:
- **Vanishing Gradient Problem:**
  - Deeper networks led to performance decline due to weakened gradient signals.
  
- **ResNet Solution:**
  - Introduced skip connections, enabling gradients to travel shorter routes.
  
- **Result:**
  - ResNet achieved superhuman performance in ImageNet classification.
  
  
  



#### 4. Availability in Keras:
- **Importing Architectures in Keras:**
  - ResNet 50, VGG 16, VGG 19, Inception, and Exception architectures are accessible in Keras.
