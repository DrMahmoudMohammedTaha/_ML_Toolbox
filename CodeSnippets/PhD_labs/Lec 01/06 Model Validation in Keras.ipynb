### Optional Resource

-   There are many callbacks (such as ModelCheckpoint) that you can use
    to monitor your model during the training process. If you'd like,
    check out the <a href="https://keras.io/callbacks/#modelcheckpoint"
    target="_blank">details</a> here. You're encouraged to begin with
    learning more about the EarlyStopping callback. If you'd like to see
    another code example of ModelCheckpoint, check out <a
    href="http://machinelearningmastery.com/check-point-deep-learning-models-keras/"
    target="_blank">this blog</a>.
